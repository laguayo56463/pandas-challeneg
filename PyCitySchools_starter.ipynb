```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
school_data_complete["school_name"].unique()
```

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
# Calculate the total number of students
studentCount = school_data_complete['Student ID'].count()
studentCount
```

```python
# Calculate the total budget
totalBudget = school_data["budget"].sum()
totalBudget
```

```python
# Calculate the average (mean) math score
avgMathScore = school_data_complete["math_score"].mean()
avgMathScore
```

```python
# Calculate the average (mean) reading score
avgReadingScore = school_data_complete["reading_score"].mean()
avgReadingScore
```

```python
#use conditional to finde students with scorew of 70 or higher
passMathStudents = school_data_complete[school_data_complete["math_score"] >=70]
#passMathStudents.head()
numPassMath = len(passMathStudents)
numPassMath
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
#number of students wo passed math / total number of students * 100
percentagePassedMath = (numPassMath / studentCount)*100
percentagePassedMath
```

```python
#use conditional to find students with score of 70 or higher
passReadingStudents = school_data_complete[school_data_complete["reading_score"] >=70]
#passReadingStudents.head()
numPassReading = len(passReadingStudents)
numPassReading
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated)  
percentagePassedReading = (numPassReading / studentCount)*100
percentagePassedReading
```

```python
# Use a condtional to find wgo passed both 
passingStudents = school_data_complete[(school_data_complete["math_score"] >=70) & (school_data_complete["reading_score"] >= 70 )]
passingStudents.head()
numPassingStudents = len(passingStudents)
numPassingStudents
```

```python
# Use the following to calculate the percentage of students that passed math and reading
percentagePassed = (numPassingStudents / studentCount)*100
percentagePassed
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
districtSummary = pd.DataFrame(
{
    "Total Schools": [school_count],
    "Total Students": [studentCount],
    "Total Budget": [totalBudget],
    "Average Math Score": [avgMathScore],
    "Avgerage Reading Score": [avgMathScore],
    "% Passing Math": [percentagePassedMath ],
    "% Passing Reading": [percentagePassedReading],
    "% Overall Passing": [percentagePassed]
})

# Formatting
districtSummary["Total Students"] = districtSummary["Total Students"].map("{:,}".format)
districtSummary["Total Budget"] = districtSummary["Total Budget"].map("{:,.2f}".format)
districtSummary["Average Math Score"] = districtSummary["Average Math Score"].map("{:,.2f}".format)
districtSummary["Avgerage Reading Score"] = districtSummary["Avgerage Reading Score"].map("{:,.2f}".format)
districtSummary["% Passing Math"] = districtSummary["% Passing Math"].map("{:,.2f}".format)
districtSummary["% Passing Reading"] = districtSummary["% Passing Reading"].map("{:,.2f}".format)
districtSummary["% Overall Passing"] = districtSummary["% Overall Passing"].map("{:,.2f}".format)


# Display the DataFrame
districtSummary
```
